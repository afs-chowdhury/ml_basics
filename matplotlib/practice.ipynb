```python
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn-whitegrid")
```

```python
# adding data to dataframe 

heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# selecting the data onlyover 50 age 

over_50 = heart_disease[heart_disease['age']>50]
over_50.head() 
```

```python
over_50["age"].min()
```

```python
fig , (ax_chol, ax_thalach) = plt.subplots(figsize=(10,20),
                      nrows=2,
                      ncols=1,
                    sharex=True)

scatter_chol = ax_chol.scatter(x=over_50["age"],
                              y=over_50["chol"],
                              c =over_50["target"],
                              cmap="winter")


ax_chol.legend(*scatter_chol.legend_elements(),title = "Target");

ax_chol.set(title="Heart disease paitents cholestrol measerement over 50 years age !",
           ylabel="Cholestrol")


ax_chol.axhline(y=over_50["chol"].mean(),
               c="r",
               linestyle="--",
               linewidth="5",
               label="mean line")


scatter_thalach = ax_thalach.scatter(x=over_50["age"],
                              y=over_50["thalach"],
                              c =over_50["target"],
                              cmap="winter")


ax_thalach.legend(*scatter_thalach.legend_elements(),title = "Target");


ax_thalach.axhline(y=over_50["thalach"].mean(),
                  c="m",
                  linestyle="--",
                  linewidth="5",
                  label="mean line")

ax_thalach.set(title="Heart disease paitents maximum heart rate  measerement over 50 years age !",
           ylabel="Maximum heart rate",
              xlabel="Age",
              xlim=[50,80],
              ylim=[50,200]);


fig.suptitle("Heart Disease Analysis",
            fontsize="20",
            fontweight="bold");
```
