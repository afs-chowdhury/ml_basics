# Methods of plotting 

```python
# first method   " long version of plt.plot(); " 

fig = plt.figure() # creates a figure
ax = fig.add_subplot()# add some axes
ax.plot(x,y) # adding some data 
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot()

ax1.plot(x,y);

ax2 = fig.add_subplot()

ax2.scatter(x,y)

plt.show()
```

```python
# second method 
fig = plt.figure() # creates figure
ax = fig.add_axes([1,1,1,1]) # creating axes . The dimensions [left, bottom, width, height] of the new axes. All quantities are in fractions of figure width and height
ax.plot(x,y)  # addig some data 
plt.show() 
```

```python
# 3rd method ( recommended )
fig, ax = plt.subplots()
ax.plot(x,y);
type(fig) , type(ax)
```

# Matplotlib example workflow 

```python
# 0. import matplotlib and get it ready for plotting in Jupyter 

%matplotlib inline 
import matplotlib.pyplot as plt

# 1 .Prepare the data 
x = [2,4,6,8]
y = [22,43,35,61]

# 2. Setup plot 

fig, ax = plt.subplots(figsize=(10,10)) # (width , height)  of the entire figure 

# 3. Plot data 

ax.plot(x,y)

# 4. Customize the plot 

ax.set(title= "Simple Plot / Figure",
      xlabel = "x-axis",
      ylabel = "y-axis")

# 5. Save yout plot 

fig.savefig("simple-plot.png")
```

## Which one should be used ? ( pyplot vs matplotlib Object oriented  method ) 

* when plotting something quickly ,  it's okay to use pyplot method 
* but when plotting something more advanced , use oo method 

```python
# Let's do some data analysis 

# selecting data of the paitents only over the age 50 

over_50 = heart_disease[heart_disease['age']>50]
over_50.head()
```

```python
len(over_50)
```

now plotting a scatterplot using the <b>pyplot</b> regarding the age and cholestrol and having heart disease or not which is specified by the target attribute as color 

```python
over_50.plot.scatter(x="age",
                    y="chol",
                    c="target",
                    figsize=(10,6));
```

```python
# now plotting with oo method mixed with pyplot method 

fig , ax = plt.subplots(figsize=(10,6))

over_50.plot.scatter(x="age",
                    y="chol",
                    c="target",
                    ax =ax);

```

```python
fig , ax = plt.subplots(figsize=(10,6))

over_50.plot.scatter(x="age",
                    y="chol",
                    c="target",
                    ax =ax);

#ax.set_xlim(left=50,right=80);

ax.set(title = " Heart disease detection based on cholestrol over age 50 ",
xlim= [50,80])

#ax.set_xlim([50,100]);
```

```python
# OO method from scratch 
fig , ax = plt.subplots(figsize=(10,6))

# plot the data 

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])


# customize the plot 

ax.set(title="Heart Disease And Cholestrol According to age over 50 year  ",
      xlabel="Age",
      ylabel="Cholestrol")

# add a legend 

ax.legend(*scatter.legend_elements(),title="Heart Disease ");
# type(scatter)

# adding a horizontal line at the mean 

ax.axhline(y=over_50["chol"].mean(),
          linestyle="--",
          c="g",
          label = "Average Line");
```

```python
over_50.head()
```

```python
# creating subplots using oo api of matplotlib

# creating figure and axes for  ax with chol and ax with maximum heartrate 

fig, (ax_chol,ax_thalach) = plt.subplots(nrows=2,
                                       ncols=1)
 
                                         #figsize=(15,20))


# adding data to ax_chol 

chol_scatter = ax_chol.scatter(x=over_50["age"],
                               y=over_50["chol"],
                               c=over_50["target"]);

# agging info of the graph 


ax_chol.set(title="Heart disease based on Measurement of cholestrol on paitents over age 50 years ",
      xlabel="Age",
      ylabel="Cholestrol");

# ading a legend 

ax_chol.legend(*chol_scatter.legend_elements(),title="Heart Diesease");

# adding a horizontal line in mean of the cholestrol value 

ax_chol.axhline(y=over_50["chol"].mean(),linestyle="--");

# adding data to ax_thalach

thalach_scatter =ax_thalach.scatter(x=over_50["age"],
                                   y=over_50["thalach"],
                 
                                    c=over_50["target"])

# adding info to the graph  


ax_thalach.set(title="Heart disease based on Maximum heart rate on  paitents over age 50 years ",
              xlabel="Age",
              ylabel="Maximum Heart Rate")

# adding leegnd 

ax_thalach.legend(*thalach_scatter.legend_elements() , title="Heart Disease");

# adding a horizontal line on the mean of maximum heart rate 

ax_thalach.axhline(y=over_50["thalach"].mean(),linestyle="--");
```

```python

# creating subplots using oo api of matplotlib

# creating figure and axes for  ax with chol and ax with maximum heartrate 

fig, (ax_chol,ax_thalach) = plt.subplots(nrows=2,
                                       ncols=1,
                                         figsize=(15,20),
                                        sharex=True)


# adding data to ax_chol 

chol_scatter = ax_chol.scatter(x=over_50["age"],
                               y=over_50["chol"],
                               c=over_50["target"]);

# agging info of the graph 


ax_chol.set(title="Heart disease based on Measurement of cholestrol on paitents over age 50 years ",
      ylabel="Cholestrol");

# ading a legend 

ax_chol.legend(*chol_scatter.legend_elements(),title="Heart Diesease");

# adding a horizontal line in mean of the cholestrol value 

ax_chol.axhline(y=over_50["chol"].mean(),linestyle="--");

# adding data to ax_thalach

thalach_scatter =ax_thalach.scatter(x=over_50["age"],
                                   y=over_50["thalach"],
                 
                                    c=over_50["target"])

# adding info to the graph  


ax_thalach.set(title="Heart disease based on Maximum heart rate on  paitents over age 50 years ",
              xlabel="Age",
              ylabel="Maximum Heart Rate")

# adding leegnd 

ax_thalach.legend(*thalach_scatter.legend_elements() , title="Heart Disease");

# adding a horizontal line on the mean of maximum heart rate 

ax_thalach.axhline(y=over_50["thalach"].mean(),linestyle="--");

# adding the whole figure title 

fig.suptitle("Heart Diesease Analysis",fontsize=16,fontweight="bold");
```

## Adding color scheme 

```python
# choosing style ( theme )

plt.style.use("seaborn-whitegrid")

# OO method from scratch 
fig , ax = plt.subplots(figsize=(10,6))

# plot the data 

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter") # adds the color scheme 


# customize the plot 

ax.set(title="Heart Disease And Cholestrol According to age over 50 year  ",
      xlabel="Age",
      ylabel="Cholestrol")

# add a legend 

ax.legend(*scatter.legend_elements(),title="Heart Disease ");
# type(scatter)

# adding a horizontal line at the mean 

ax.axhline(y=over_50["chol"].mean(),
          linestyle="--");
```

```python
# setting themes or style to the plot 

plt.style.use("seaborn-whitegrid")

# creating subplots using oo api of matplotlib

# creating figure and axes for  ax with chol and ax with maximum heartrate 

fig, (ax_chol,ax_thalach) = plt.subplots(nrows=2,
                                       ncols=1,
                                         figsize=(10,10),
                                        sharex=True)


# adding data to ax_chol 

chol_scatter = ax_chol.scatter(x=over_50["age"],
                               y=over_50["chol"],
                               c=over_50["target"],
                              cmap="winter"); # setting up colour scheme 

# agging info of the graph 


ax_chol.set(title="Heart disease based on Measurement of cholestrol on paitents over age 50 years ",
      ylabel="Cholestrol");

# ading a legend 

ax_chol.legend(*chol_scatter.legend_elements(),title="Heart Diesease");

# setting x limits 
ax_chol.set_xlim([50,80])

# adding a horizontal line in mean of the cholestrol value 

ax_chol.axhline(y=over_50["chol"].mean(),linestyle="--");

# adding data to ax_thalach

thalach_scatter =ax_thalach.scatter(x=over_50["age"],
                                   y=over_50["thalach"],
                 
                                    c=over_50["target"],
                                   cmap= "winter")

# adding info to the graph  


ax_thalach.set(title="Heart disease based on Maximum heart rate on  paitents over age 50 years ",
              xlabel="Age",
              ylabel="Maximum Heart Rate",
              xlim=[50,80],
              ylim=[50,200])  # another way of setting xlim and ylim 

# adding leegnd 

ax_thalach.legend(*thalach_scatter.legend_elements() , title="Heart Disease");

# adding a horizontal line on the mean of maximum heart rate 

ax_thalach.axhline(y=over_50["thalach"].mean(),linestyle="--");

# adding the whole figure title 

fig.suptitle("Heart Diesease Analysis",fontsize=16,fontweight="bold");
```
